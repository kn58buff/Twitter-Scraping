```python
subset = pd.read_csv("data/validation/SUP.csv").drop(columns = ["index"])
main = pd.read_csv("data/raw_tweets/target_data.csv")
```

```python
def join_info(main, subset):
    """ Add corresponding date and country information to tweet dataframe

    :param main: Main dataframe where date information is obtained
    :param subset: Subset of main dataframe, where date/country info will be added
    :return: DataFrame with column "Date" prepended and "Country" appended
    """

    # Retrieve corresponding dates
    dates = [] # Container for found dates
    for text in subset["Text"]:
        dates.append(main.loc[main["Text"] == text]["Date"].values[0])
    dates_df = pd.DataFrame(dates, columns = ["Date"]) # Convert to df

    # Identify corresponding countries
    countries = [] # Container for found countries
    with open("data/countries.csv", "r") as countriesfile:
        reader = csv.reader(countriesfile)
        next(reader)
        for row in reader:
            countries.append(row)
    countries.remove(["KE", "Kenya"]) # Remove kenya from countries list

    # Same iteration as before, pretty much
    rel_countries = []
    for text in subset["Text"]:
        for country in countries:
            if country[1] in text: # check if country is in the tweet
                rel_countries.append(country[1])
    countries_df = pd.DataFrame(rel_countries, columns = ["Country"]) # Convert to dataframe
    merged_df = pd.concat([dates_df, subset, countries_df], axis = 1) # Join all 3 dataframes into one

    return merged_df
```
